### list (révisions - vues au TP 4)

Une liste est **modifiable** ("mutable") et **ordonnée** (indicée par des entiers à partir de 0), elle peut contenir des **données hétérogènes et dupliquées**. On peut créer une liste avec **[]** ou **list()**.

* Créez ci-dessous une **variable l** contenant une liste contenant les 100 premiers nombres pairs (entiers positifs, 0 est pair). Puis affichez là.

### tuple (liste non modifiable)

Un tuple est **NON modifiable** ("immutable") et **ordonnée** (indicée par des entiers à partir de 0), elle peut contenir des **données hétérogènes et dupliquées**. On peut créer une liste avec **()** ou **tuple()**.

* Créez ci-dessous un **tuple t** contenant les 50 premiers nombres entiers impairs.

_(aide : passez par une list comprehension que vous convertirez en tuple grâce à la fonction **tuple()**)_

```python
li=[(elt*2)+1 for elt in range(50)]
t=tuple(li)
t
```

Comme indiqué en cours les **tuples sont _non modifiables_** :
    
    * Vous pouvez utiliser les mêmes méthodes pour accéder aux données/éléments que les listes
    * Mais les méthodes pour modifier un tuple sont interdites, par exemple :
        * Pas le droit de faire t[0] = 10 
        * Pas le droit d'utiliser la méthode append(), t.append(10) par exemple
        
Si vous allez modifier votre liste : utiliser une liste

Mais si vous voulez une liste que vous n'allez pas modifier, alors utiliser un tuple. Les accès aux données seront alors plus rapides.

### set (ensemble)

Un set (ensemble) est **modifiable** ("mutable"). Il n'est **PAS ordonné**. Il peut contenir des **données hétérogènes** mais **NON dupliquées** : chaque donnée d'un set est donc **unique**. On peut créer un set avec **set()**.

Les <strong>sets (ensembles)</strong> sont donc optimisés pour contenir des données <strong>non ordonnées</strong> et <strong>sans doublons</strong> (un même objet ne peut être présent qu'une seule fois dans l'ensemble) afin de réaliser des opérations ensemblistes :

    * Intersection avec le symbole &
    * Union avec le symbole |
    * Différence avec le symbole -
    

Considérons deux variables **s1** et **s2** contenant deux sets :

```python
s1 = {1,2,3,4}
s2 = {3,4,"a","b"}
```

* Affichez s1 et s2 

```python
print(s1)
print(s2)
```

_(Remarquez que l'ordre peut différer de la saisie)_

* Affichez :
    * l'**intersection** de <code>s1</code> et <code>s2</code> (Les éléments présents dans <code>s1</code> et <code>s2</code>)
    * l'**union** de <code>s1</code> et <code>s2</code> (Les éléments présents dans <code>s1</code> ou <code>s2</code>)
    * la **différence** <code>s1-s2</code> (Les éléments de <code>s1</code> qui ne sont pas dans <code>s2</code>)
    * la **différence** <code>s2-s1</code> (Les éléments de <code>s2</code> qui ne sont pas dans <code>s1</code>)

(La différence n'est pas une opération commutative)

```python
print(f"Intersection : {s1&s2}")
print(f"Union : {s1|s2}")
print(f"s1 moins s2 : {s1-s2}")
print(f"s2 moins s1 : {s2-s1}")
```
